```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Path 
base_path = "D:/PROJECT/Pizza_Project/data/"

# CSV Files Load with (Encoding)
orders = pd.read_csv(base_path + 'orders.csv')
order_details = pd.read_csv(base_path + 'order_details.csv')
pizzas = pd.read_csv(base_path + 'pizzas.csv')

pizza_types = pd.read_csv(base_path + 'pizza_types.csv', encoding='latin-1')

# Merge The Data
df = pd.merge(order_details, pizzas, on='pizza_id')
df = pd.merge(df, pizza_types, on='pizza_type_id')

print("✅ Data Loaded Successfully!")
```

```python
# Group  Data
top_pizzas = df.groupby('name')['quantity'].sum().sort_values(ascending=False).head(5)

#  Graph 
plt.figure(figsize=(10, 6))

sns.barplot(x=top_pizzas.values, y=top_pizzas.index, hue=top_pizzas.index, legend=False, palette='viridis')

plt.title('Top 5 Best Selling Pizzas', fontsize=16)
plt.xlabel('Total Quantity Sold')
plt.ylabel('Pizza Name')
plt.show()
```

```python
# Time to hours
# Use 'orders'  table
orders['hour'] = orders['time'].apply(lambda x: int(str(x).split(':')[0]))

# 2. Graph  (Peak Time)
plt.figure(figsize=(12, 6))
# 'color' 
sns.countplot(x='hour', data=orders, color='orange')

# 3. Label 
plt.title('Busiest Hours of the Day (Peak Time)', fontsize=16)
plt.xlabel('Hour (24 Hr Format)')
plt.ylabel('Total Orders')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```
